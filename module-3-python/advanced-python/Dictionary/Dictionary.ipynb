```python
employee = {1:"brijesh",2:"krish",3:"Het",4:"Divyang"}
#itretion key
for k in employee:
    print(k)
```

```python
#itretion velue
for k in employee.values():
    print(k)
```

```python
for i in employee: 
    print(employee[i])
```

```python
emp=dict({1:"brijesh",2:"krish",3:"Het",4:"Divyang"})
print(emp[1])
```

```python
print(emp.get(1))
print(emp.get(4))
print(emp.get(2))
print(emp.get(3))
```

```python
#Nested Dictionary
fruit={1:"banana",2:"apple",3:{"a":"pineapple","b":"guvava"}}
print(fruit)
print(fruit[2])
print(fruit[3])
print(fruit[3]["b"])
```

```python
# Add new item in dictionary 
print(emp)
emp[7]="uouytre"
print(emp)
```

```python
# Update a Dictionary
emp[5]="Raju"
print(emp)
```

```python
#Delete 
del emp[3]
print(emp)
```

```python
for Key,Value in emp.items():
    print(f"{Key}:{Value}")
```
